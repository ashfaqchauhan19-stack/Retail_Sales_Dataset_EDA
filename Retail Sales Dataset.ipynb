### 6. Data visualization

```python
data.Gender.value_counts().plot(kind='pie',autopct='%.2f%%')
plt.ylabel(" ")                             
```

#### In the above insights shows that there are more no of females as compared to males.

```python
data['Product Category'].value_counts()
```

```python
sns.countplot(data=data, x='Product Category')
plt.xticks(rotation=50)
plt.show()
```

#### In the above insights shows that clothing and electronics products have more no of sales.

```python
sns.histplot(data['Total Amount'], kde=True)
plt.show()
```

#### In the above insights shows that distribution is right-skewed (few large purchases)

```python
data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
```
